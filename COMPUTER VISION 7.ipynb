```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread(r"dogpark.jpg", cv2.IMREAD_GRAYSCALE)

# Prewitt kernels for edge detection
prewitt_kernel_x = np.array([[ -1, 0, 1],
                             [ -1, 0, 1],
                             [ -1, 0, 1]])

prewitt_kernel_y = np.array([[ -1, -1, -1],
                             [ 0,  0,  0],
                             [ 1,  1,  1]])

# Apply the Prewitt operator to compute gradients
gradient_x = cv2.filter2D(image, -1, prewitt_kernel_x)
gradient_y = cv2.filter2D(image, -1, prewitt_kernel_y)

# Compute the magnitude of the gradient
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# Normalize the magnitude to range [0, 255] for display
gradient_magnitude = np.uint8(np.clip(gradient_magnitude, 0, 255))


normalized_gradient_magnitude = cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX)

# Apply Canny edge detection on the normalized gradient
canny_edges = cv2.Canny(normalized_gradient_magnitude, 50, 150)  # Adjust thresholds as necessary

# Plotting the results
plt.figure(figsize=(10, 7))

# Original image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Prewitt edge detection result
plt.subplot(1, 3, 2)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Prewitt Edge Detection')
plt.axis('off')

# Canny edge detection result
plt.subplot(1, 3, 3)
plt.imshow(canny_edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread(r"dogpark.jpg", cv2.IMREAD_GRAYSCALE)

# Define the Roberts Cross operator kernels
roberts_kernel_x = np.array([[1, 0],
                              [0, -1]])

roberts_kernel_y = np.array([[0, 1],
                              [-1, 0]])

# Apply the Roberts operator to compute gradients
gradient_x = cv2.filter2D(image, -1, roberts_kernel_x)
gradient_y = cv2.filter2D(image, -1, roberts_kernel_y)

# Compute the magnitude of the gradient
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# Normalize the magnitude to range [0, 255] for display
gradient_magnitude = np.uint8(np.clip(gradient_magnitude, 0, 255))

# Normalize the gradient magnitude for better visualization
normalized_gradient_magnitude = cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX)

# Apply Canny edge detection on the normalized gradient
canny_edges = cv2.Canny(normalized_gradient_magnitude, 50, 150)  # Adjust thresholds as necessary

# Plotting the results
plt.figure(figsize=(10, 7))

# Original image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Roberts edge detection result
plt.subplot(1, 3, 2)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Roberts Edge Detection')
plt.axis('off')

# Canny edge detection result
plt.subplot(1, 3, 3)
plt.imshow(canny_edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')

plt.show()
```
